#### Bounds of the grids to obtain

worst case: calculate 1000 km (~8-9 degrees lat) around the TC center => lat range (8S, 55N), lon range (53E, 164W) under assumption of Mercator projection
longitude ranges are incorrect but we need a rectangular grid, so...

```python
records = ds.query("(SEASON > 1991) or ((SEASON == 1991) and (ISO_TIME_MONTH > 7))")
records["USA_LAT"].describe()
# latitude range for dynamical data: use range must have 1 to 46
```

```python
nearby_lon = records.query("(USA_LON >= 50) and (USA_LON <= 180)")
nearby_lon["USA_LON"].describe()
# longitudes must include 62
```

```python
faraway_lon = records.query("(USA_LON <= -100)")
faraway_lon["USA_LON"].describe()
# longitudes must include -175
# a very large range, but it is necessary
```
